```python
import os
import matplotlib.pyplot as plt
import imageio
import numpy as np

from skimage import measure
from mpl_toolkits.mplot3d.art3d import Poly3DCollection
import matplotlib.cm as cm



# function to visualize the 3D volume starting from 2D mask slices
def plot_3d(image, threshold=-300): 
    p = image.transpose(2,1,0)
    verts, faces, normals, values = measure.marching_cubes_lewiner(p, threshold)
    fig = plt.figure(figsize=(20, 20))
    ax = fig.add_subplot(111, projection='3d')
    
    mesh = Poly3DCollection(verts[faces], alpha=0.6)
    
    face_color = [0.5, 0.5, 1]
    mesh.set_facecolor(face_color)
    ax.add_collection3d(mesh)
    ax.set_xlim(0, p.shape[0])
    ax.set_ylim(0, p.shape[1])
    ax.set_zlim(-5, 12)
    plt.savefig("esempio_segmentation_3d.png")
    plt.show()
```

```python
path_images = "Data/Train/Images/Sunnybrook_Part1/"
path_labels = "Data/Train/Labels/Sunnybrook_Part1/"

# select all the slices of a single patient
list_names = [
"SC-HF-I-02-IM-0001-0060.dcm.png",
"SC-HF-I-02-IM-0001-0080.dcm.png",
"SC-HF-I-02-IM-0001-0100.dcm.png",
"SC-HF-I-02-IM-0001-0120.dcm.png",
"SC-HF-I-02-IM-0001-0140.dcm.png",
"SC-HF-I-02-IM-0001-0160.dcm.png",
"SC-HF-I-02-IM-0001-0180.dcm.png",
"SC-HF-I-02-IM-0001-0200.dcm.png",
"SC-HF-I-02-IM-0001-0220.dcm.png"]
```

```python
images_list = []
labels_list = []

# read images
for file in list_names:
    image = imageio.imread(os.path.join(path_images, file))
    label = imageio.imread(os.path.join(path_labels, file))

    image = image[...,0][...,None]/255

    label = label[...,0]>1
    label = label[...,None]
    label = label.astype(np.int64)

    images_list.append(image)
    labels_list.append(label)
    
# in images_list we have all the slice images of the patient
images_list = np.array(images_list)
# in labels_list we have all the masks of the patient
labels_list = np.array(labels_list)
```

```python
# we can plot the volume also in this way

labels_list_reshape = np.transpose(labels_list[:,:,:,0], (1,2,0))
```

```python
x = np.zeros(shape = (256,256,9))
y = np.zeros(shape = (256,256,9))
z = np.zeros(shape = (256,256,9))

for i in range(256):
    x[i,:,:] = i
    y[:,i,:] = i
    
for k in range(9):
    z[:,:,k] = k
```

```python
cm.register_cmap(name='alpha_gradient',
                 data={'red':   [(0.,0,0),
                                 (1.,0,0)],

                       'green': [(0.,0.6,0.6),
                                 (1.,0.6,0.6)],

                       'blue':  [(0.,0.4,0.4),
                                 (1.,0.4,0.4)],

                       'alpha': [(0.,0,0),
                                 (1.,0.2,0.2)]})
```

```python
fig = plt.figure(figsize=(10,10))
ax = fig.add_subplot(111, projection='3d', azim = +60)

k = 1
p = ax.scatter(x[80:150,80:150,::k], y[80:150,80:150,::k], z[80:150,80:150,::k],c=images_list_reshape[80:150,80:150,::k], cmap='alpha_gradient')
ax.set_zlim(-3,12)
fig.colorbar(p)
```
